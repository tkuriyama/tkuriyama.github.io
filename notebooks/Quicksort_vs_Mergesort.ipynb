```python
# Euler-Mascheroni constant
gamma = 0.577215664901532
```

```python
def qsort_cmp(n):
    return (2 * n * np.log(n) - 
            2 * (1-gamma) * n)
```

```python
def msort_cmp(n):
    return (n * np.floor(np.log2(n)) - 
            2 ** (np.floor(np.log2(n))+1) + 
            2 * n)
```

```python
qsort_cmp(100000)
```

```python
msort_cmp(100000)
```

<hr>

# QuickSort vs MergeSort

```python
x1 = range(1, 10**10, 100000)
y1 = [qsort_cmp(x) for x in x1]
```

```python
x2 = x1[:]
y2 = [msort_cmp(x) for x in x2]
```

```python
plt.plot(x1, y1, label = "Quicksort")
plt.plot(x2, y2, label = "Mergesort")
plt.xlabel('Input')
plt.ylabel('Comparisons')
plt.title('Comparison Count')
plt.legend()
plt.show()
```

# MergeSort Terms

```python
def msort_cmp_term1(n):
    return n*np.floor(np.log2(n))
```

```python
def msort_cmp_term2(n):
    return 2**(np.floor(np.log2(n))+1)
```

```python
def msort_cmp_term3(n):
    return 2*n
```

```python
mx = range(1, 2**8+2)
my = [msort_cmp(x) for x in mx]
my1 = [msort_cmp_term1(x) for x in mx]
my2 = [msort_cmp_term2(x) for x in mx]
my3 = [msort_cmp_term3(x) for x in mx]
```

```python
plt.plot(mx, my, label="Mergesort")
plt.plot(mx, my1, label="Term 1")
plt.plot(mx, my2, label="Term 2")
plt.plot(mx, my3, label="Term 3")
plt.xlabel('Input')
plt.ylabel('Comparisons')
plt.title('Mergesort # of Comparisons')
plt.legend()
plt.show()
```
